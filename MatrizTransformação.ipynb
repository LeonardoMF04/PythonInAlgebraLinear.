```python
#Rotacionar a imagem para que ela fique o mais reto possvel

import cv2 # <- est  a openCV
import numpy as np
import matplotlib.pyplot as plt
# lendo a imagem
image = cv2.imread("quadro.jfif")
# mudando a imagem para rgb
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
# definindo a transformao
M = np.float32([
    [np.cos(np.pi/10), -np.sin(np.pi/10), 100],
    [np.sin(np.pi/10),  np.cos(np.pi/5), -100]])
# guardando as dimenses da imagem
linhas, colunas, camadas = image.shape
# aplicando a transformao
image = cv2.warpAffine(image, M, (colunas, linhas))
# mostrando imagem
im = plt.imshow(image, interpolation="none")
plt.axis("off")
plt.show()

cv2.imwrite('imagem_corrigida.jpg', img)
```
