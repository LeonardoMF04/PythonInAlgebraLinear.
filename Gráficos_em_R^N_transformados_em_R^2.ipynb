```python
import pandas as pd
import numpy as np  
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# Coletar ou obter o conjunto de dados
df = pd.read_csv("accelerometer.csv")

# Separar os dados em atributos e rótulos
X1 = df[["wconfid", "pctid", "x", "y"]]
Y1 = df["z"]

# Realizar a redução de dimensão com PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X1)

# Visualizar os dados reduzidos com PCA
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=Y1)
plt.xlabel("Componente 1")
plt.ylabel("Componente 2")
plt.title("Visualização dos dados reduzidos com PCA")
plt.show()
```
